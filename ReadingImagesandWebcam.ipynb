function to capture live image from webcam

0 for built in camera
1-... for external camera connceted

```python
cap = cv2.VideoCapture(0)
```

setting the dimensions of the frame

the output is a booleran value

```python
cap.set(3, 640)
cap.set(4, 480)
cap.set(10,100)
```

as the camera send a live stream of many frames continoucontinously, so here is a function to process multiple frames

using while loop to conticontinuously read from the camera

when the live stream will be there, u may notice that it might be flipped, so to flip the image, we use flip function
parameters are video and 1(it is the axis)

to stop the loop

we use waitkey and a special keypress to terminate the program

then we release all the cap varibale and destroy all windows 

```python
while True:
  success, vid = cap.read()
  vid = cv2.flip(vid, 1)
  cv2.imshow("Video", vid)
  if cv2.waitKey(1) & 0xFF == ord('q'):
    break

cap.release()
cv2.destroyAllWindows()

```
